```python
import os

dataset_path = 'train'

for class_name in os.listdir(dataset_path):
    class_folder = os.path.join(dataset_path, class_name)
    if os.path.isdir(class_folder):
        num_images = len([
            file for file in os.listdir(class_folder)
            if file.lower().endswith(('.jpg', '.jpeg', '.png', '.bmp'))
        ])
        print(f"{class_name}: {num_images} images")
```

```python
import os
import random

dataset_path = 'train(2)'

# Step 1: Count how many images in each folder
class_counts = {}
for class_name in os.listdir(dataset_path):
    class_folder = os.path.join(dataset_path, class_name)
    if os.path.isdir(class_folder):
        images = [f for f in os.listdir(class_folder)
                  if f.lower().endswith(('.jpg', '.jpeg', '.png', '.bmp'))]
        class_counts[class_name] = len(images)

# Step 2: Find minimum class count
min_count = min(class_counts.values())
print(f"Target image count per class: {min_count}")

# Step 3: Randomly delete images from classes with more than min_count
for class_name, count in class_counts.items():
    if count > min_count:
        class_folder = os.path.join(dataset_path, class_name)
        images = [f for f in os.listdir(class_folder)
                  if f.lower().endswith(('.jpg', '.jpeg', '.png', '.bmp'))]
        to_delete = random.sample(images, count - min_count)
        
        for file_name in to_delete:
            file_path = os.path.join(class_folder, file_name)
            os.remove(file_path)
            print(f"Deleted: {file_path}")

print(" Image balancing complete.")
```
